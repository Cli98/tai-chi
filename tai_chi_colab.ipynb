## Open up a dataframe
> This is the most of your coding work here

```python
df = pd.read_csv('./sample_data/california_housing_train.csv')
```

## Start Tai-Chi Engine 🚀 ☯️

```python
# set a location where the prject is going to end
PROJECT = Path("./housing")

# start the engine
engine = TaiChiEngine(df, project = PROJECT)
engine
```

```python
engine()
```

## Model inference
> After the training, let's load the model for inference

```python
from tai_chi_engine import TaiChiTrained
```

```python
trained = TaiChiTrained(PROJECT)
trained
```

```python
trained.predict({"longitude":-119.17, "median_income":3.6500})
```
